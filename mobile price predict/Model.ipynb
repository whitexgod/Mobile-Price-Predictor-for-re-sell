```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
df=pd.read_csv('mobile_prices.csv')
```

```python
df.shape
```

```python
df
```

```python
df['Average_price']=(df['Highest_price']+df['lowest_price'])/2
```

```python
df.sample(20)
```

```python
y=df['Average_price'].values
```

```python
y.shape
```

```python
df.drop(columns=['Highest_price','lowest_price','Average_price','Unnamed: 0'],inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df['brand'].value_counts()
```

```python
df['rom']=df['rom'].astype('category')
```

```python
df['ram']=df['ram'].astype('category')
```

```python
df.info()
```

```python
df
```

```python
front_cam=[]
for i in df['Front_cam']:
    try:
        if int(i)<=50:
            front_cam.append(i)
        else:
            front_cam.append(np.nan)
    except:
        front_cam.append(np.nan)
```

```python
df['front_cam']=front_cam
```

```python
df.drop(columns='Front_cam',inplace=True)
```

```python
df=df.replace(np.nan,5)
```

```python
df
```

```python
df['front_cam']=df['front_cam'].astype('float')
```

```python
df.info()
```

```python
df.drop(columns='Model_name',inplace=True)
```

```python
df['ram']=df['ram'].astype('object')
df['rom']=df['rom'].astype('object')
```

```python
df.info()
```

```python
final=pd.get_dummies(df, columns=['brand','ram','rom'])
```

```python
final
```

```python
final.columns.values
```

```python
X=final.iloc[:,:].values
```

```python
X
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print('The accuracy of the model = ',r2_score(y_test,y_pred)*100)
```
