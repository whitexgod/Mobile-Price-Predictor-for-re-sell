```python
import requests
import bs4
import pandas as pd
```

```python
src=requests.get('https://www.phonecurry.com/benchmarks')
```

```python
soup = bs4.BeautifulSoup(src.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
box = soup.find('table', class_='table table-condensed benchmarks-table')
```

```python
name=[]
score=[]
for each_candidate in box.find_all('tbody'):
    rows = each_candidate.find_all('tr')
    for row in rows:
        name.append(row.find_all('td')[1].text.strip())
        score.append(row.find_all('td')[2].text.strip())
```

```python
len(name)
```

```python
len(score)
```

```python
d={'processor':name, 'performance':score}
df=pd.DataFrame(d)
```

```python
df
```

```python
df['performance'].value_counts()
```

```python
cost=[]
for i in range(df.shape[0]):
    if df['performance'].iloc[i]=='Super High':
        cost.append(1500)
    elif df['performance'].iloc[i]=='High':
        cost.append(1000)
    elif df['performance'].iloc[i]=='Moderate':
        cost.append(500)
    else:
        cost.append(300)
```

```python
df['cost']=cost
```

```python
df
```

```python
df.to_csv('processor_details.csv')
```
